## Using python to graph linear equations

#### What are linear equations?

-linear equations are equations that are literally a "line" when graphed.

The standard formula for linear equations are:

$$ y=mx+b$$

where m = slope and b = the y-intercept.

$$ slope = \frac{\Delta y}{\Delta x}$$

In python, the most common graphing library is `matplotlib`

```python
import matplotlib.pyplot as plt


### generate a linear equation using :  y = 5x + 2


##y_values = []

### generate x_values for our plot
x_values = list(range(0,15))


### now evaluate the y_values, using y = 5x + 2

"""for x in x_values:
    y_values.append(5*x + 2)
"""
y_values = list(map(lambda x: 5*x +2, x_values))

#### Now use matplotlib to generate the plot. syntax --> plt.plot(x,y)

plt.plot(x_values,y_values) # REQUIRED
plt.ylabel('y') ## optional
plt.xlabel('x')  ## optional
plt.title('Plot of y = 5x + 2') ## optional
plt.show  # REQUIRED
```

### We can use python to find the slope, intercept, and print the equation of a line when given x and y values.


```python
import matplotlib.pyplot as plt


x_values = [-2, -1, 0, 1, 2]  ## x
y_values = [-7, -4, -1, 2, 5]  ## y(x)


slope = (y_values[3] - y_values[2])/(x_values[3] - x_values[2])  ## slope = (y2-y1)/(x2-y2)

## to find the intercept, we have to solve for b using the slope

### solving for b in the equation gives b = y - mx

b = y_values[1] - slope*x_values[1]

### Now we can ask python to print the equation, the slope, the intercepr, and graph it.

print("The equation of the line is: y = {}x + {}".format(slope,b))
print("The slope is {}".format(slope))
print("The y-intercept is {}".format(b))

plt.plot(x_values,y_values)
plt.grid(True)
plt.show()
```

#### You can write functions that compute the slope and intercept given x,y values so that you don't have to keep writing the same thing over and over.

```python
x_values = [-2, -1, 0, 1, 2]  ## x
y_values = [-7, -4, -1, 2, 5]  ## y(x)

def slope_of(x,y):

    slope = (y[-1] - y[0])/(x[-1] - x[0]) ## This uses the last element and first element of a the lists
    return slope


def find_b(m,x,y):
    ## b = y - mx
    b = y[0] - m*x[0]
    return b


slope = slope_of(x_values, y_values)

y_intercept = find_b(slope, x_values,y_values)

print(slope)
print(y_intercept)
    
```
