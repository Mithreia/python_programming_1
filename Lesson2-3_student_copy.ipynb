# Lesson 3: Python Data Types - Part 1

### There are 4 basic data types in Python:

- Int
- Float
- String
- Boolean

**Int**
: This is an integer number. For example 1, 2, 3, 40, 465, etc.

**Float**
: This is used for "floating point numbers", which are numbers with decimals.

**String**
: This is used for what we consider "words" 

**Boolean**
: These are used for logic operations. Can be either True or False.
